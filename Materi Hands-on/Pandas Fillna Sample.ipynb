```python
import numpy as np
import pandas as pd
```

a dictionary

```python
data_dict = {
    'col_a' : [1, np.nan, 3, np.nan, np.nan, 5],
    'col_b' : ['a', 'b', 'c', np.nan, np.nan, 'd'],
    'col_c' : [1, None, 3, 4, np.nan, 5],
    'col_d' : [1.2, 2.2, 3.2, np.nan, np.nan, 5.]
}
```

create a dataframe from a dictionary

```python
df = pd.DataFrame(data_dict)
df
```

cek apakah ada missing value ? 

```python
df.isna().sum()
```

pada script berikut ini, kita lakukan proses replace column dengan missing value dengan nilai baris berikutnya yang tidak kosong. terlihat untuk baris-baris data yang kosong, akan diisi dengan nilai dari baris setelahnya. yang kita lakukan pada script di bawah ini, kita hanya melakukan preview saja. data df nya sendiri tidak berubah. jika anda ingin mengisi kolom dengan missing value dengan nilai dari kolom sebelumnya yang ada nilai, gunakan method='ffill'. anda juga bisa mengisi kolom dengan missing value dengan sebuah nilai scalar, atau bisa jg dari fungsi aggregate seperti mean, modus, median.

```python
df.fillna(method='bfill')
```

penggunaan parameter threshold pada fungsi dropna ?

```python
df = df_temp.copy()
df
```

```python
df.dropna(how='any', thresh=2)
```

pada script di atas, pada penggunaan parameter thresh, parameter ini akan menyeleksi jika sebuah baris data (default jika tidak ditulis parameter axis, default axis=0, yang artinya by rows) harus minimal memiliki kolom nilai yang tidak missing sebanyak nilai pada thresh agar dipertahankan. jika baris tersebut tidak memiliki nilai kolom dengan nilai sebanyak thresh, maka baris tersebut akan di drop. pada contoh di atas, baris dengan minimal ada 2 nilai ada pada baris dengan index 0, 1, 2, 5. baris-baris ini akan dipertahankan. sedangkan yang lain akan di drop.

bagaimana jika kita ganti nilai tresh menjadi 3 ? 
maka baris yang akan dipertahankan menjadi baris dengan index 0, 2, dan 5.

```python
df.dropna(how='any', thresh=3)
```
