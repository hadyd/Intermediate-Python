### 5. Get rid of the missing value:
#### a. Drop columns which has missing value more than 50% of the total data
#### b. Fill the missing value in agent with the mean
#### c. Fill the missing value in country, children with the most common namea

```python
# a. Drop columns which has missing value more than 50% of the total data
data_hotel.dropna(thresh=0.5*len(data_hotel), axis=1, inplace=False)
```

```python
# b. Fill the missing value in agent with the mean
data_hotel.loc[:, "agent"].fillna(data_hotel.loc[:, "agent"].mean())
```

```python
# c. Fill the missing value in country, children with the most common name
data_hotel.loc[:, "country"].fillna(data_hotel.loc[:, "country"].mode()[0])
```

```python
data_hotel.loc[:, "children"].fillna(data_hotel.loc[:, "children"].mode()[0])
```

### 6. Make category based on the lead time: Short, Medium, and Long

```python
data_hotel["category"] = pd.cut(data_hotel["lead_time"], bins=3, labels=["Short", "Medium", "Long"])
data_hotel.loc[:, ["lead_time", "category"]]
```

### 8. Visualize the arrival_date_month to see in what month the most visitors come

```python
visualize = data_hotel["arrival_date_month"].value_counts()
visualize
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
visitors = [0]*len(months)
```

```python
for month in months:
    idx = months.index(month)
    visitors[idx] = visualize[month]
visitors
```

```python
plt.barh(months, visitors, color='y')
plt.ylabel("Month", fontdict = {"fontsize": 18})
plt.xlabel("Number of Visitors", fontdict = {"fontsize": 18})
plt.show()
```
