Answer 1:

```python
def is_prime(n):
    if n % 2 == 0 and n > 2 or (n == 0 or n == 1) : 
        return False
    return all(n % i for i in range(3, int(math.sqrt(n)) + 1, 2))

array = np.arange(1, 26).reshape(5,5)
foo = np.vectorize(is_prime)
pbools = foo(array)
primes = np.extract(pbools, array)


print("Array yang dimiliki (sebelum difilter):") 
print(array) 
print("----------------------------") 
print("Filtering mendapatkan bilangan prima:") 
print(primes)
```

Answer 2:

```python
def isPrime(n):
    if n == 1:
        return False
    if n == 2:
        return True
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
arr = np.arange(1, 26).reshape(5, 5)
arr
```

```python
prima = []
for line in arr:
    for item in line:
        if isPrime(item):
            prima.append(item)
            
arrPrime = np.array(prima)
arrPrime
```

```python
arrPrime2 = np.array([item for item in arr.flatten() if isPrime(item)])
arrPrime2
```
